```python
import math
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
# You can just run all of these at the beginning apparently and it's chill with that.

x = sp.symbols('x')     # defines x as a variable
a = 7                   # 7 is the last number of my student ID
f = 1/sp.sqrt(a*x + 1)   # defines 'that' as f

sp.plot(f,(x,0,1))         # graphing it from 0-1
```

### Based on the graph (and my knowledge of the function), the line starts at y=1 and ends a little less than y=0.4. Plugging 1 into the function for x gives me 1/sqrt(8), which is around 1/2.8. So a little more than a half is what the area should come out to be.

```python
area = sp.integrate(f, (x, 0, 1))   # This makes 'area' be the integral of f from 0 to 1

print("Area from 0 -> 1 =")
display(area)
print("As an exact number")
print(float(area))
```
